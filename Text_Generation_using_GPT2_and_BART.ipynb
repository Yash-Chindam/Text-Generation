```python
template = """Question: Tell me abt {question}

Answer: Let's think step by step."""
prompt = PromptTemplate.from_template(template)
```

```python
model_id="gpt2"
model=AutoModelForCausalLM.from_pretrained(model_id)
tokenizer=AutoTokenizer.from_pretrained(model_id)
pipe=pipeline("text-generation",model=model,tokenizer=tokenizer,max_new_tokens=100,model_kwargs={"prompt":prompt,"max_new_tokens": 100})
hf=HuggingFacePipeline(pipeline=pipe)
```

```python
question=input("Name the topic u want to know abt  ")
print(hf.invoke(question))
```
